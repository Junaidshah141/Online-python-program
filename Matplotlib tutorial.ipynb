```python
from matplotlib import style
```

```python
Languages = ['Urdu','english', 'Balochi', 'Hindi', 'Pashto']
prv_sindh = [40,11,30,10,9]
prv_punjab = [22,24,17,7,20]
prv_sarhad = [34,11,15,27,13]
```

```python
style.use('ggplot')
plt.figure(figsize=(14,7))
plt.bar(Languages, prv_sindh, align = 'edge', width = 0.4, color = 'g')
plt.xlabel("Languages", fontsize = 18)
plt.ylabel("No. of speaker", fontsize = 18)
plt.title("Speakers", fontsize = 18)
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(14,7))

class_index = np.arange(len(Languages))
width = 0.3
plt.bar(class_index, prv_sindh,  width, color = "g", label = "Sindh")
plt.bar(class_index + width, prv_punjab,  width, color = "r", label ="Punjab")
plt.bar(class_index + width +width, prv_sarhad,  width, color = "m", label = "Khyber")
plt.xticks(class_index + width, Languages, rotation = 90)

plt.xlabel("Languages", fontsize = 18)
plt.ylabel("No. of speaker", fontsize = 18)
plt.title("Speakers", fontsize = 18)
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.figure(figsize=(14,7))

class_index = np.arange(len(Languages))
width = 0.3
plt.barh(class_index, prv_sindh,  width, color = "g", label = "Sindh")
plt.barh(class_index + width, prv_punjab,  width, color = "r", label ="Punjab")
plt.barh(class_index + width +width, prv_sarhad,  width, color = "m", label = "Khyber")
plt.yticks(class_index + width, Languages)

plt.ylabel("Languages", fontsize = 18)
plt.xlabel("No. of speaker", fontsize = 18)
plt.title("Speakers", fontsize = 18)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


data = {'Barton LLC': 109438.50,
        'Frami, Hills and Schmidt': 103569.59,
        'Fritsch, Russel and Anderson': 112214.71,
        'Jerde-Hilpert': 112591.43,
        'Keeling LLC': 100934.30,
        'Koepp Ltd': 103660.54,
        'Kulas Inc': 137351.96,
        'Trantow-Barrows': 123381.38,
        'White-Trantow': 135841.99,
        'Will LLC': 104437.60}
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)
```

```python
fig, ax = plt.subplots()

ax.bar(group_names,group_data)
```

```python
plt.style.use('fivethirtyeight')
```

```python
fig, ax = plt.subplots(figsize=(16,9))

ax.barh(group_names,group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation = 90)
```

```python
plt.rcParams.update({'figure.autolayout': True})

fig, ax = plt.subplots(figsize=(14,8))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
```

```python
def currency(x, pos):
    """The two arguments are the value and tick position"""
    if x >= 1e6:
        s = '${:1.1f}M'.format(x*1e-6)
    else:
        s = '${:1.0f}K'.format(x*1e-3)
    return s
```

```python
plt.rcParams.update({'figure.autolayout': True})

fig, ax = plt.subplots(figsize=(14,8))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set( xlabel = 'Revenue', ylabel='Company', title = 'Companies Revenue')
ax.xaxis.set_major_formatter(currency)
```

```python
fig, ax = plt.subplots(figsize=(14, 8))
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')

# Add a vertical line, here we set the style in the function call
ax.axvline(group_mean, ls='--', color='r')

# Annotate new companies
for group in [3, 5, 8]:
    ax.text(145000, group, "New Company", fontsize=10,
            verticalalignment="center")

# Now we move our title up since it's getting a little cramped
ax.title.set(y=1.05)

ax.set(xlim=[-10000, 140000], xlabel='Total Revenue', ylabel='Company',
       title='Company Revenue')
ax.xaxis.set_major_formatter(currency)
ax.set_xticks([0, 25e3, 50e3, 75e3, 100e3, 125e3])
fig.subplots_adjust(right=.1)

plt.show()
```
