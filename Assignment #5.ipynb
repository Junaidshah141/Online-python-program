```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv("netflix_titles.csv")
df.head(5)
```

```python
df.loc[:, ["type"]].value_counts()
```

```python
df["cast"] = df["cast"].replace(np.nan, 0)
#arr = df.iloc[1, df["cast"]].split(", ")
#arr


df['no_of_cast'] = df.cast.apply(lambda x: len(str(x).split(' ')))
df.head(5)
```

```python
df.head(10)
```

```python
data = df.groupby(["type", "rating", "no_of_cast"])['country'].value_counts()
```

```python
data.to_csv('types.csv')
```
